```python
device = 'cuda:0'
conf = ffhq256_autoenc()
# print(conf.name)
model = LitModel(conf)
state = torch.load(f'/home/anthony/experiments/diffae/ipace_autoenc/epoch=4-step=112463.ckpt', map_location='cpu')
model.load_state_dict(state['state_dict'], strict=True)
model.ema_model.eval()
model.ema_model.to(device);
```

```python
data = ImageDataset('imgs_align', image_size=conf.img_size, exts=['jpg', 'JPG', 'png'], do_augment=False)
batch = data[1]['img'][None]
```

```python
from torchvision import transforms
image_size = 256
transform = [
            transforms.Resize(image_size),
            transforms.CenterCrop(image_size),
        ]
transform.append(transforms.ToTensor())
transform.append(
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)))
transform = transforms.Compose(transform)
```

```python
from PIL import Image
img = Image.open('/home/anthony/datasets/dino_train/2019-12-18--15-24-08--Nicolas--way-back/1576682782037939unixus.jpeg')

img = transform(img)

batch = img[None]
```

```python
import matplotlib.pyplot as plt
plt.imshow(batch[0].permute([1, 2, 0]) / 2 + 0.5)
```

# Encode

```python
cond = model.encode(batch.to(device))
xT = model.encode_stochastic(batch.to(device), cond, T=250)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ori = (batch + 1) / 2
ax[0].imshow(ori[0].permute(1, 2, 0).cpu())
ax[1].imshow(xT[0].permute(1, 2, 0).cpu())
```

# Decode
